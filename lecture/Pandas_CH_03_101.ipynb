## Working with Funcion

```python
def salary_type(salary):
  if salary>50000:
    return 'high'
  else:
    return 'enough'
```

```python
df["salary"].apply(salary_type)
```

```python
df["salary"].apply(lambda x: salary_type(x))
```

```python
df["salary"].apply(lambda x: "high" if x>50000 else "enough")
```
